**-------------------------Q7---------------------**

If a five-digit number is input through the keyboard, write a program to reverse the number

```python
n = eval(input('Enter a five digit number:'))
rev = (n % 10) * 10000 \
    + (n // 10 % 10) * 1000 \
    + (n // 100 % 10) * 100 \
    + (n // 1000 % 10) * 10 \
    + (n // 10000)
print(f'Reversal of number {n} is: {rev}')
```

**-------------------------Q9---------------------**

In a town, the percentage of men is 52. The percentage of total literacy is 48. If total percentage of literate
men is 35 of the total population, write a program to find the total number of illiterate men and women if
the population of the town is 80,000.

```python
total_population = 80000
total_literacy_population = total_population * (48/100)
man_population = total_population * (52/100)
woman_population = (total_population) - (man_population) 
man_literate_population = total_population * (35/100)
woman_literate_population = total_literacy_population - man_literate_population
man_illiterate = man_population - man_literate_population
woman_illiterate = woman_population - woman_literate_population
print (f'Total men population are {man_population}')
print (f'Total wommen population are {woman_population}')

print (f'Total literate men are {man_literate_population}')
print (f'Total literate women are {woman_literate_population}')

print (f'Total illiterate men are {man_illiterate}')
print (f'Total illiterate women are {woman_illiterate}')
```
