*wap ask the user take two number print subtraction multiplication and division*

```python
num1 = eval(input('Enter first number:'))
num2 = eval(input('Enter second number:'))
add = num1 + num2
print (f'The addition of {num1} and {num2} is {add}')
sub = num1 - num2
print (f'The subtraction of {num1} and {num2} is {sub}')
mul = num1 * num2
print (f'The subtraction of {num1} and {num2} is {mul}')
div = num1 / num2
print (f'The subtraction of {num1} and {num2} is {div}')
```

```python
radius = eval(input('Enter the radius of the circle'))
import math
area = round(math.pi * (math.pow(radius,2)),2)
print(f'The area of circle is {area}')
```


**wap ask the user take the salary, percetage of tax calculate tax amount**


```python
Base_salary = eval(input('Base salary is Rs/-'))
if Base_salary > 0 and Base_salary <= 400000:
    tax_percentage = 0
elif Base_salary > 400000 and Base_salary <= 800000:
    tax_percentage = 5
elif Base_salary > 800000 and Base_salary <= 1200000:
    tax_percentage = 10
elif Base_salary > 1200000 and Base_salary <= 1600000:
    tax_percentage = 15
elif Base_salary > 1600000 and Base_salary <= 2000000:
    tax_percentage = 20
elif Base_salary > 2000000 and Base_salary <= 2400000:
    tax_percentage = 25
else:
    tax_percentage = 30

tax_amount = Base_salary * (tax_percentage/100)
print(f'The tax is {tax_percentage}%')
print(f'The tax amount is {tax_amount}')

         
```
